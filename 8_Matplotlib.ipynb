# Matplotlib 

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+

 * Generally easy to get started
 * Support for custom labels and texts
 * Great control of every element in a figure
 * High-quality output in many formats
 * Very customizable in general

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
## using this will not require to use plt.show()
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(50,60)
b=np.arange(60,70)
```

```python
# Plotting using matplotlib

plt.scatter(x,y,c='g') # g is green colour
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
# plot

plt.plot(x,y,'r*')
```

```python
plt.plot(x,y,'ro-')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r-*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
# subplots

plt.subplot(2,2,1)
plt.plot(x,y,'ro-')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'b--')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form')
plt.plot(x,y)
```

```python
# subplots

x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)

plt.plot(x,y_sin,'r--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

plt.show()
```

```python
# barplot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x, y)
plt.bar(x2, y2, color='r')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Histograms

```python
a=np.array([22,41,56,24,78,65,48,86,34,68,72,98,11,17])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

## Box Plot 

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True);
```

```python
plt.boxplot(data,vert=True,patch_artist=False);
```

```python
plt.boxplot(data,vert=False,patch_artist=False);
```

## Pie Chart

```python
labels='Python','C++','Ruby','Java'
sized=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0.2,0,0) # explode 1st slice

plt.pie(sized,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
